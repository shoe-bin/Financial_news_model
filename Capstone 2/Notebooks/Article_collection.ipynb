```python
class Benzinga:
    
    def __init__(self):
        self.div = ["sc-cVuPSX", "cwFmwT", "article-content-body-only"]
        self.soup = None
        self.session = requests.Session()
        
    def create_soup(self, url):
        response = self.session.get(url)
        self.soup = BeautifulSoup(response.content, 'html.parser')
        
    def get_article(self):
        article = ''
        div_tag = self.soup.find('div', class_ = self.div)
        p_tags = div_tag.find_all('p')
        for p in p_tags:
            article += p.get_text() + ' '
        return article
            
    def fit(self, url):
        self.create_soup(url)
        article = self.get_article()
        return article
    
class Zack:
    
    def __init__(self):
        self.div = ["commentary_body"]
        self.header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)\
        AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'}
        self.soup = None
        self.session = requests.Session()
        
    def create_soup(self, url):
        response = self.session.get(url, headers=self.header)
        self.soup = BeautifulSoup(response.content, 'html.parser')
        
    def get_article(self):
        article = ''
        div_tag = self.soup.find('div', class_ = self.div)
        p_tags = div_tag.find_all('p')
        for p in p_tags:
            article += p.get_text() + ' '
        return article
            
    def fit(self, url):
        self.create_soup(url)
        article = self.get_article()
        return article
    
class Fool:
    
    def __init__(self):
        self.div = ["article-body"]
        self.soup = None
        self.session = requests.Session()
        
    def create_soup(self, url):
        response = self.session.get(url)
        self.soup = BeautifulSoup(response.content, 'html.parser')
        
    def get_article(self):
        article = ''
        div_tag = self.soup.find('div', class_ = self.div)
        p_tags = div_tag.find_all('p')
        for p in p_tags:
            article += p.get_text() + ' '
        return article
            
    def fit(self, url):
        self.create_soup(url)
        article = self.get_article()
        return article


class Investor:
    
    def __init__(self):
        self.div = ["single-post-content", "post-content", "drop-cap crawler"]
        self.soup = None
        self.session = requests.Session()
        
    def create_soup(self, url):
        response = self.session.get(url)
        self.soup = BeautifulSoup(response.content, 'html.parser')
        
    def get_article(self):
        article = ''
        div_tag = self.soup.find('div', class_ = self.div)
        p_tags = div_tag.find_all('p')
        for p in p_tags:
            article += p.get_text() + ' '
        return article
            
    def fit(self, url):
        self.create_soup(url)
        article = self.get_article()
        return article
```
